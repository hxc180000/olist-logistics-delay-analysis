```python
df_category = pd.read_csv("../data_notes/delay_by_category.csv")

df_category.head()
```

```python
top10 = df_category[["product_category", "delay_rate_percent"]].copy()

top10["product_category"] = top10["product_category"].replace("nan", pd.NA)
top10 = top10.dropna(subset=["product_category", "delay_rate_percent"])
top10 = top10.head(10)
```

```python
plt.figure(figsize=(10,6))

plt.barh(top10["product_category"].tolist(), top10["delay_rate_percent"].tolist())

plt.xlabel("Delay Rate (%)")
plt.ylabel("Product Category")
plt.title("Top 10 Product Categories with Highest Delivery Delay")

plt.gca().invert_yaxis()

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.barh(top10["product_category"].tolist(), top10["delay_rate_percent"].tolist())

plt.xlabel("Delay Rate (%)")
plt.ylabel("Product Category")
plt.title("Top 10 Product Categories with Highest Delivery Delay")

plt.gca().invert_yaxis()

plt.tight_layout()
plt.savefig("../charts/delay_by_category.png", dpi=150)

plt.show()
```
